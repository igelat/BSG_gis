```python
# importing geopandas
import geopandas as pdg
```

```python
# reading filename
file_name = "ugandadistricts.geojson"
data = pdg.read_file(file_name)
```

```python
# display data
data.head(130)
```

```python
# access shape area
data["shape_area"]=data.Shape_Area
```

```python
#setting an index
data=data.set_index("Subregion")
data["shape_Area"]=data.shape_area
```
